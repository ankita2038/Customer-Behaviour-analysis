Feature Engineering

```python
# Create Total Spending (optional, already available in dataset)
df['Total Spending'] = df['Total Purchase Amount']

# Create Purchase Month
df['Purchase Month'] = df['Purchase Date'].dt.month

# Create Purchase Year
df['Purchase Year'] = df['Purchase Date'].dt.year

df.head()
```

Create RFM Table

```python
import datetime as dt

snapshot_date = df['Purchase Date'].max() + dt.timedelta(days=1)

rfm = df.groupby('Customer ID').agg({
    'Purchase Date': lambda x: (snapshot_date - x.max()).days,
    'Customer ID': 'count',
    'Total Purchase Amount': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

rfm.head()
```

```python
rfm.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

rfm_scaled = scaler.fit_transform(rfm)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

rfm.head()
```

```python
rfm['Cluster'].value_counts()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=rfm,
    x='Frequency',
    y='Monetary',
    hue='Cluster',
    palette='Set2'
)

plt.title("Customer Segmentation using KMeans")

plt.show()
```

```python
rfm.groupby('Cluster').mean()
```

Cluster 0:
High Value Customers

Cluster 1:
Regular Customers

Cluster 2:
Low Spending Customers

Cluster 3:
Potential Churn Customers
